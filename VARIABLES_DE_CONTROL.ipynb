# Variables de control

Importamos la libreria y definimos N y el valor real de la integral de la variable de control

```python
import numpy as np


N = 1000000
mu_h = 0.75
```

Definimos la función a usar, así como las variables de control y el vallr exacto de la integral

```python
def g(x):
  return 1.0 / (1.0 + x**2)

def h(x):
  return 1.0 - 0.5 * x

Valor_Real = np.pi / 4.0
print(f"Valor Real de la Integral: {Valor_Real:.8f}\n")
```

Simulamls por Montevarlo crudo, y sacamos varianza y error

```python
np.random.seed(42)
X = np.random.uniform(0, 1, N)

Y_g = g(X)

I_MC_crudo = np.mean(Y_g)

Var_MC_crudo = np.var(Y_g) / N  # Varianza de la estimación I_MC
DE_MC_crudo = np.sqrt(Var_MC_crudo)  # Desviación Estándar de la estimación

print("--- Monte Carlo Crudo ---")
print(f"Estimación de I: {I_MC_crudo:.8f}")
print(f"Varianza de la Estimación: {Var_MC_crudo:.8e}")
print(f"Error Absoluto: {np.abs(I_MC_crudo - Valor_Real):.8f}")
```

Simulamos con variables de control

```python
Y_h = h(X)

Cov_gh = np.cov(Y_g, Y_h)[0, 1] # Covarianza entre g(X) y h(X)
Var_h = np.var(Y_h)             # Varianza de h(X)

c_opt = -Cov_gh / Var_h

Z = Y_g + c_opt * (Y_h - mu_h)
I_MC_control = np.mean(Z)

Var_MC_control = np.var(Z) / N # Varianza de la estimación I_VC
DE_MC_control = np.sqrt(Var_MC_control) # Desviación Estándar de la estimación

print("\n--- Monte Carlo con Variables de Control ---")
print(f"Factor de control óptimo (c*): {c_opt:.4f}")
print(f"Estimación de I: {I_MC_control:.8f}")
print(f"Varianza de la Estimación: {Var_MC_control:.8e}")
print(f"Error Absoluto: {np.abs(I_MC_control - Valor_Real):.8f}")
```

Posteriormente comparamos la reduccion de varianza

```python
Reduccion_Varianza = (Var_MC_crudo - Var_MC_control) / Var_MC_crudo
Porcentaje_Reduccion = Reduccion_Varianza * 100

print("\n--- Comparación ---")
print(f"Reducción de Varianza de la Estimación: {Porcentaje_Reduccion:.2f}%")
```
