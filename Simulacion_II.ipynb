```python
import tabulate
def generador(a,b,m,x0):
    x1=(a*x0+b) %m
    return x1
n=64
a=5
b=7
m=64
x0=1

muestra=[]
datos=[]

for i in range(n):
    nuevo=generador(a,b,m,x0)
    muestra.append([i,x0,nuevo/64])
    datos.append(x0/64)
    x0=nuevo

encabezados=["i", "x_i", "X_{i+1}"]
print(tabulate.tabulate(muestra, headers=encabezados,tablefmt="psql",numalign="center"))
print(datos)
```
